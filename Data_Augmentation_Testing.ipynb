```python
import torch
import torchvision
import os
import random
import flow_transforms_new
from presets import OpticalFlowPresetTrain, OpticalFlowPresetEval
```

```python
# With Data Augmentation
fc_transforms = OpticalFlowPresetTrain(crop_size=(368, 496), min_scale=0.1, max_scale=1.0, do_flip=False)
flyingChairs = torchvision.datasets.FlyingChairs(root="../FlyingChairs", split="train", transforms=fc_transforms)

img1 = torchvision.transforms.ToPILImage()(flyingChairs[20][0])
img2 = torchvision.transforms.ToPILImage()(flyingChairs[20][1])
flow = torchvision.transforms.ToPILImage()(torchvision.utils.flow_to_image(flyingChairs[20][2]))
```

```python
# With Data Augmentation
img1.show()
img2.show()
flow.show()
```
